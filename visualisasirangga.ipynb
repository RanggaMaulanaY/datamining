```
owners = '1st owner'
data[data['owner']==owners]
```

```
unique_name = data[data.year>2008]['name'].unique()
unique_name.sort
unique_name
```

```
motocycle_bledosan = []
for i in unique_name:
  motocycle_bledosan.append(data[data.year>0][data['name']==i][data['owner']==owners].year.sum())
```
